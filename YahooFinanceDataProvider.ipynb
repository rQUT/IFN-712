```python
import datetime
import matplotlib.pyplot as plt
from qiskit_finance import QiskitFinanceError
from qiskit_finance.data_providers import YahooDataProvider

try:
    # Fetch data using YahooDataProvider
    data = YahooDataProvider(
        tickers=["NVDA", "NIO", "LCID", "TSLA"],
        start=datetime.datetime(2022, 1, 1),
        end=datetime.datetime(2023, 12, 31),
    )
    data.run()

    # Plot data for each ticker
    plt.figure(figsize=(12, 6))
    for (cnt, s) in enumerate(data._tickers):
        plt.plot(data._data[cnt], label=s)

    # Adding labels and title
    plt.xlabel("Date")
    plt.ylabel("Stock Price")
    plt.title("Stock Price Trend from 2022 to 2023 for Selected Tickers", pad=20)  # Adding padding to the title
    
    
    plt.legend(loc="upper left", bbox_to_anchor=(1, 1), ncol=1)

    # Formatting x-axis
    plt.xticks(rotation=90)
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.tight_layout()  # Automatically adjust subplots to fit in figure area
    plt.show()

except QiskitFinanceError as ex:
    data = None
    print(ex)
```
